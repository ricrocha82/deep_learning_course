```python
# separate price from the other features
X = df.drop(['price',"date"], axis =1).values
Y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.30, random_state=101) 
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# create a sequential model
X_train.shape
# 19 features = 19 neurons ]
```

```python
model = Sequential()

# add 4 layers
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

# output layer
model.add(Dense(1))

# add optimizer
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
# fit the model with validation data
# fit in batches (to avoid memory issues)
model.fit(x=X_train, y=y_train, 
            validation_data=(X_test, y_test), 
            batch_size=128, epochs=400)
```

```python
# model evaluation
# convert the history into a dataframe
pd.DataFrame(model.history.history)
```

```python
# we can plot the history
# train vs validaiton loss
# epoch as x-axis
# if the lines are close to each other, it means that the model is not overfitting
pd.DataFrame(model.history.history).plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predicitons = model.predict(X_test)
```

```python
np.sqrt(mean_absolute_error(y_test, predicitons))
```

```python
mean_absolute_error(y_test, predicitons)
```

```python
df['price'].describe()['mean']
```

```python
mean_absolute_error(y_test, predicitons) / df['price'].describe()['mean']
# around 18% error, which is not bad but not great
```

```python
explained_variance_score(y_test, predicitons)
# the variance is 80%, which is good but can be better
```

Use reshape(1, 19):

When you explicitly need a 2D array with exactly 1 row and 19 columns.

When you donâ€™t know or care about the number of rows but want each row to have 19 columns. Let NumPy figure out the row count.
