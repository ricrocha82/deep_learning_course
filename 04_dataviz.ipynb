```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# if jupyter NB doesn't show the plot just type
# %matplotlib inline
import seaborn as sns
```

## matplot

Matplot is a basic tool to make basic plots. Just for a quick visualization. 

```python
# a simple plot
x = [0,1,2]
y = [100,200,300]
```

```python
plt.plot(x,y)
```

```python
# add semicolon to hide matplotlib text output
plt.plot(x,y);
```

```python
# For running .py files!
plt.plot(x,y)
plt.show()
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                       'price':[100,120,190,200,230,310,330,305]})
housing
```

```python
# scatter plot
plt.scatter(housing['rooms'], housing['price']);
```

```python
# add lables and titles
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y label');
```

```python
# change the plot
# change colors
# add markers
plt.plot(x,y, color = 'red', marker = 'o', markersize = 15, linestyle = '--')

# add limits
plt.xlim(0,2) 
plt.ylim(0,300)

# add title and lables
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y label');
```
