# Part One - DataFrames

```python
# create an array of number with numpy
np.random.seed(123)
arr = np.random.randint(-100,100,(5,4)) # a matrix with 5 rows and 4 columns
arr
```

```python
# we need 4 columns
columns = ["A", 'B', 'C', 'D']
index = ['E', 'F', 'G', 'H', 'I'] # row names
```

```python
# creating the Dataframe
df = pd.DataFrame(arr, index, columns)
df
```

## Columns

```python
df['A'] # return a Series/column
```

```python
# to select more than one column
# need to pass a list
df[["A", 'B']]
```

```python
# how to create a new column?
df['new_column'] = df['C'] + df['D']
df
```

```python
# remove a column
df = df.drop('new_column', axis = 1) # defaul is axis=0, so row. Need to change to axis=1 (column)
df
```

```python
df.iloc[4]
df.iloc[-1] # the very last row
df.iloc[0:2]
```

```python
# but if selecting more than one row, it isn't transpose the data
df.loc[['E', 'F']]
```

```python
# dropping a row
df.drop("F") # axis=0 (row) is the default
```

```python
# locate values
df.loc['E', "A"]
```

```python
# locate values in more than one columns or rows
df.loc[["E", 'G'], 'B']
```

We are going to select based on conditions

```python
df > 0 # return booleans values
```

```python
df[df > 0]
```

let's check only in one column/series

```python
df['C'] > 0  # only the rows F and G have values > 0 ...
```

```python
df[df['C'] > 0] # ... so, if we pass this condition into the df, it wil only select the rows F and G
# and will return all the df
```

Multiple condiitons

`&` and `|`

```python
# two conditions
# need to be inside of paranthesis
(df['C'] > 0) & (df['D'] > 1)
```

```python
(df['C'] > 0) | (df['D'] > 1)
```

```python
# put all those things into the df
df[(df['C'] > 0) & (df['D'] > 1)]
```

```python
df[(df['C'] > 0) | (df['D'] > 1)]
```

# Missing data

Let NA as NA

Remove NA, or

Replace it?

```python
df = pd.DataFrame({'A':[1,2, np.nan,4],
                   'B':[5, np.nan, np.nan,8],
                   'C':[10,20,30,40]})
```

```python
df
```

```python
# removing NAs
df.dropna() # defual is rows
```

```python
df.dropna(axis = 1) # columns
```

```python
df.dropna(axis = 1, thresh = 2) # add a threshold (non-NAs)
```

```python
df.dropna(axis = 1, thresh = 3) # remove columns with more than 2 non-NAs
# can add a percentage for example
# .25*len(df)
```

```python
# filling NAs
df.fillna(value = "FILL VALUE")
```

```python
df.fillna(value = 0)
```

```python
# fill only one column
df['A'].fillna(value = 'NA')
```

```python
df['A'] = df["A"].fillna(value = 1)
df
```

```python
# filling NA with mean
df['B'].fillna(value = df['B'].mean())
```

```python
df['B'] = df['B'].fillna(value = df['B'].mean())
df
```

```python
# or can do it throughout the entire df
df = pd.DataFrame({'A':[1,2, np.nan,4],
                   'B':[5, np.nan, np.nan,8],
                   'C':[10,20,30,40]})
df
```

```python
df.fillna(value = df.mean())
```

## apply

applying functions to the dataframe

```python
# first define a function
def grab_first_letter(state):
  return state[0]
```

```python
grab_first_letter('NY')
```

```python
df_one['first_letter'] = df_one['col2'].apply(grab_first_letter) # I didin't use () after the function
df_one
```

```python
# another function
def complex_letter(state):

  if state[0] == "W":
    return "Washington"
  else:
    return state[0]
```

```python
df_one['col2'].apply(complex_letter)
```

## map

to use map, the keys should be the same in the columns that you want to map to

```python
# first let's build a dictionary to map
my_map = {'A':1, "B":2, "C":3}
my_map
```

```python
df_one['numbers'] = df_one['k1'].map(my_map)
```

```python
# check the max value
df_one['col1'].max()
```

```python
# where is located?
df_one['col1'].idxmax()
```

```python
# the minimum value
def minimum_value(df, col):
  min_value = df[col].min()
  index_min = df[col].idxmin()
  return print(f"the minimum value is {min_value}, located at {index_min}")

minimum_value(df_one, 'col1')
```

## concatenating DFs

binding them by rows or column (axis)

```python
features = pd.DataFrame({'A':[100,200,300,400,500],
                        'B':[12,13,14,15,16]})
predictions = pd.DataFrame({'pred':[0,1,1,0,1]})
```

```python
# binding rows
pd.concat([features, predictions], axis = 0)
```

```python
# binding cols
pd.concat([features, predictions], axis = 1)
```
